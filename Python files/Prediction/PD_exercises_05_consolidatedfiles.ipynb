```python
import numpy as np
import pandas as pd
import math
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('C:/Users/Nguyen/Documents/Python Scripts/fclaims_pastyr.csv')
data.drop("MemberID", axis = 1, inplace = True)
data.drop("Year", axis = 1, inplace = True)

y = data['Target']
X = data.loc[:, data.columns != 'TARGET']
X.head()
```

```python
def eval(model, X_test, y_test):
    y_hat = model.predict(X_test) # your predicted y values
    y_hat = y_hat.astype(int)
    # The root mean square logarithmic error
    msle = np.mean( (np.log(y_hat + 1) - np.log(y_test + 1)) ** 2)
    rmsle = math.sqrt(msle)
    return rmsle
```

```python
# Scale data Y1
# Chọn Standard Scaler - Chia từng value cho standard deviation
# Gần với dạng normal distribution và cũng giảm outlier

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(X)
new_X = scaler.transform(X)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(max_depth = 20, max_features = 60
                  , min_samples_leaf =  2, bootstrap = True, min_impurity_decrease = 0.0
                  , random_state = 0, n_estimators = 2000, n_jobs = -1, oob_score = False)
model.fit(new_X, y)
```

```python
eval(model, new_X, y)
```
